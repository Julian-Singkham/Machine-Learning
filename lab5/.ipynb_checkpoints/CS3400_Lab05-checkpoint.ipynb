```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
from sklearn import datasets 
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import classification_report
from neurons import Input
from neurons import Neuron
from neurons import HStack
```

### Train Multilayer Perceptron (MLP) Models on Petal and Sepal Features

```python
iris = datasets.load_iris()
scaled_X = StandardScaler().fit_transform(iris.data)

# Train Petals MLP
mlp_petals = MLPClassifier(hidden_layer_sizes=(4,), max_iter=1000, solver="lbfgs")
mlp_petals.fit(scaled_X[:, 2:], iris.target)

# Train Sepal MLP
mlp_sepal = MLPClassifier(hidden_layer_sizes=(4,), max_iter=1000, solver="lbfgs")
mlp_sepal.fit(scaled_X[:, :2], iris.target)
```

### Visualize Lines Learned by Individual Neurons

Re-arrange the following equation to solve for $x_2$
    $$0 = B_0 + B_1x_1 + B_2x_2$$  
    $$B_2x_2=-B_0-B_1x_1$$  
    $$x_2=-\frac{B_0+B_1x_1}{B_2}$$  

```python
# Extract the weight vectors for the hidden layers
# The columns of this matrix correspond to B0, B1, and B2
# Each row corresponds to the model from a separate neuron
mlp_petals_models = np.vstack([mlp_petals.intercepts_[0], mlp_petals.coefs_[0]]).T
mlp_sepal_models = np.vstack([mlp_sepal.intercepts_[0], mlp_sepal.coefs_[0]]).T
```

#### Petals

```python
# Plot features 0 and 1 from scaled_X along with the Lines for the 4 neurons in the hidden layer
x = np.linspace(-2, 2, num=100)
y1 = (mlp_petals_models[0,0]+mlp_petals_models[0,1]*x)/mlp_petals_models[0,2]
y2 = (mlp_petals_models[1,0]+mlp_petals_models[1,1]*x)/mlp_petals_models[1,2]
y3 = (mlp_petals_models[2,0]+mlp_petals_models[2,1]*x)/mlp_petals_models[2,2]
y4 = (mlp_petals_models[3,0]+mlp_petals_models[3,1]*x)/mlp_petals_models[3,2]

plt.title("Petals: Height vs Width")
plt.scatter(scaled_X[:50,0], scaled_X[:50,1], label = "setosa")
plt.scatter(scaled_X[50:100,0], scaled_X[50:100,1], label = "versicolor")
plt.scatter(scaled_X[100:,0], scaled_X[100:,1], label = "virginica")
plt.plot(x, y1, label="Neuron 1")
plt.plot(x, y2, label="Neuron 2")
plt.plot(x, y3, label="Neuron 3")
plt.plot(x, y4, label="Neuron 4")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.xlim([-2, 3])
plt.xlabel("Height", fontsize=16)
plt.ylim([-3, 4])
plt.ylabel("Width", fontsize=16)
```

#### Sepal

```python
# Plot features 0 and 1 from scaled_X along with the Lines for the 4 neurons in the hidden layer
y1 = (mlp_sepal_models[0,0]+mlp_sepal_models[0,1]*x)/mlp_sepal_models[0,2]
y2 = (mlp_sepal_models[1,0]+mlp_sepal_models[1,1]*x)/mlp_sepal_models[1,2]
y3 = (mlp_sepal_models[2,0]+mlp_sepal_models[2,1]*x)/mlp_sepal_models[2,2]
y4 = (mlp_sepal_models[3,0]+mlp_sepal_models[3,1]*x)/mlp_sepal_models[3,2]

plt.title("Sepal: Height vs Width")
plt.scatter(scaled_X[:50,2], scaled_X[:50,3], label = "setosa")
plt.scatter(scaled_X[50:100,2], scaled_X[50:100,3], label = "versicolor")
plt.scatter(scaled_X[100:,2], scaled_X[100:,3], label = "virginica")
plt.plot(x, y1, label="Neuron 1")
plt.plot(x, y2, label="Neuron 2")
plt.plot(x, y3, label="Neuron 3")
plt.plot(x, y4, label="Neuron 4")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.xlabel("Height", fontsize=16)
plt.ylim([-2, 2])
plt.ylabel("Width", fontsize=16)
```

### Visualize Outputs from ReLU Activation Function

```python
# Create a mesh grid in the range of [-2, 2] along each dimension.
x = np.linspace(-2, 2, num=20)
x1, x2 = np.meshgrid(x, x)
mesh = np.array([x1.flatten(), x2.flatten()]).T

plt.title("Mesh Grid")
plt.scatter(mesh[:,0], mesh[:,1])
```

#### Petals

```python
# Use the Input and Neuron classes to calculate the value for the hidden layer neuron at each grid point
# Pass the grid points into the predict() method as X
input = Input()
fig, axs = plt.subplots(4, 1, dpi=80, figsize=(9,14))
fig.suptitle("Petals Neuron Heatmap")
for i in range(4):
    p_layer = Neuron([input], mlp_petals_models[i, :])
    pred = p_layer.predict(mesh)
    points_heat = np.reshape(pred, [20,20]) # reshape pred to fit into heatmap
    sb.heatmap(points_heat, ax=axs[i])
    axs[i].set_title(str("Neuron "+str(i)))    
```

#### Sepal

```python
# Use the Input and Neuron classes to calculate the value for the hidden layer neuron at each grid point
# Pass the grid points into the predict() method as X
input = Input()
fig, axs = plt.subplots(4, 1, dpi=80, figsize=(9,14))
fig.suptitle("Sepal Neuron Heatmap")
for i in range(4):
    p_layer = Neuron([input], mlp_sepal_models[i, :])
    pred = p_layer.predict(mesh)
    points_heat = np.reshape(pred, [20,20]) # reshape pred to fit into heatmap
    sb.heatmap(points_heat, ax=axs[i])
    axs[i].set_title(str("Neuron "+str(i)))
```

### Train Logistic Regression models on Transformed and Original Features

#### Petals

```python
# Recreate the hidden layers 
input = Input()
p_layer_1 = Neuron([input], mlp_petals_models[0, :])
p_layer_2 = Neuron([input], mlp_petals_models[1, :])
p_layer_3 = Neuron([input], mlp_petals_models[2, :])
p_layer_4 = Neuron([input], mlp_petals_models[3, :])
stacked = HStack([p_layer_1, p_layer_2, p_layer_3, p_layer_4])

# Predict the transformed values to create a transformed feature matrix  
transformed_petals_X = stacked.predict(scaled_X[:, 2:])
```

#### Sepal

```python
# Recreate the hidden layers 
input = Input()
p_layer_1 = Neuron([input], mlp_sepal_models[0, :])
p_layer_2 = Neuron([input], mlp_sepal_models[1, :])
p_layer_3 = Neuron([input], mlp_sepal_models[2, :])
p_layer_4 = Neuron([input], mlp_sepal_models[3, :])
stacked = HStack([p_layer_1, p_layer_2, p_layer_3, p_layer_4])

# Predict the transformed values to create a transformed feature matrix  
transformed_sepal_X = stacked.predict(scaled_X[:, :2])
```

```python
#Combine the two transformed feature matrix into a new feature matrix with 8 columns
transformed_X = np.hstack((transformed_petals_X,transformed_sepal_X))
```

```python
# Logrithmic regression trained on the original 4 features
sgdc_4 = SGDClassifier(loss="log")
sgdc_4.fit(iris.data, iris.target)
pred_4 = sgdc_4.predict(iris.data)
cm = confusion_matrix(iris.target, pred_4)
print(cm) 
cr = classification_report(iris.target, pred_4)
print(cr)
```

```python
# Logrithmic regression trained on the transformed feature matrix with 8 columns
sgdc_8 = SGDClassifier(loss="log")
sgdc_8.fit(transformed_X, iris.target)
pred_8 = sgdc_8.predict(transformed_X)
cm = confusion_matrix(iris.target, pred_8)
print(cm) 
cr = classification_report(iris.target, pred_8)
print(cr)
```
