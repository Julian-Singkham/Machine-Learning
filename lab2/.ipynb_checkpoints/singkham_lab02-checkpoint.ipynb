### Experiment 1: Gaussian Distribution

#### Part I â€“ Given Data

```python
import numpy as np
import scipy
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Ingest gaussdist.csv and split the response and feature data.
np_feature = np.loadtxt(open('gaussdist.csv', "rb"), delimiter=",", usecols=range(0,1))
np_response = np.loadtxt(open('gaussdist.csv', "rb"), delimiter=",", usecols=range(1,2))

# Plot the gaussdist data
plt.title("Gaussian Distribution")
plt.scatter(np_feature, np_response)
plt.xlabel("Feature", fontsize=16)
plt.ylabel("Response", fontsize=16)
```

#### Part II â€“ Model Familiarity

Gaussian Distribution Equation  
$ f(x) = \frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{1}{2}\frac{(x-\mu)^2}{\sigma}} $

$ \mu $ = Mean  
$ \sigma $ = Standard Deviation  
x = Input  
f(x) = Output


#### Part III â€“ Model Implementation

```python
#Function that calculates the gaussian distribution
def gauss_eq(np_x, np_mu, np_sigma):
    np_results = np.zeros((np_x.size, np_mu.size))
    for j in range(0, np_mu.size):
        for i in range(0, np_x.size):
            np_results[i,j] = 1/(pow(2*np.pi,0.5)*np_sigma[j])*np.exp(-0.5*pow(np_x[i]-np_mu[j],2)/np_sigma[j])
    return np_results
```

#### Part IV â€“ Model output and Visualization

```python
# Create predictions
np_feature_predictions = gauss_eq(np_feature, np.array([3]), np.array([1]))
np_mu = np.array([1,1,5,6])
np_sigma = np.array([0.75,1.25,1.25,1.25])
np_feature_predictions_parameters = gauss_eq(np_feature, np_mu, np_sigma)

# Plot the predictions data
plt.title("Gaussian Distribution")
plt.scatter(np_feature, np_response, label="Dataset")
plt.scatter(np_feature, np_feature_predictions, label="mu=3 sigma=1")
plt.scatter(np_feature, np_feature_predictions_parameters[:,0], label="mu=1 sigma=0.75")
plt.scatter(np_feature, np_feature_predictions_parameters[:,1], label="mu=1 sigma=1.25")
plt.scatter(np_feature, np_feature_predictions_parameters[:,2], label="mu=5 sigma=1.25")
plt.scatter(np_feature, np_feature_predictions_parameters[:,3], label="mu=6 sigma=1.25")
plt.legend(loc="upper right")
plt.xlabel("Feature", fontsize=16)
plt.ylabel("Response", fontsize=16)
```

#### Part V â€“ Error between Model and Data:

Mean Absolute Error

mae = $ (\frac{1}{n})\sum_{i=1}^{n}\left | y_{i} - x_{i} \right | $

$ n $ = Number of data points  
$ y_{i} $ = True values  
$ x_{i} $ = Predicted values   
mae = The error rate. 0.0 is best

```python
# Calculate mean absolute error
np_predictions_error = np.sum((abs(np_feature_predictions_parameters.T- np_response)), axis=1)/np_response.size

# Plot the predictions data with mean absolute error
plt.title("Gaussian Distribution")
plt.scatter(np_feature, np_response, label="Dataset")
plt.scatter(np_feature, np_feature_predictions_parameters[:,0], label="mu=1 sigma=0.75 error={}".format(np_predictions_error[0]))
plt.scatter(np_feature, np_feature_predictions_parameters[:,1], label="mu=1 sigma=1.25 error={}".format(np_predictions_error[1]))
plt.scatter(np_feature, np_feature_predictions_parameters[:,2], label="mu=5 sigma=1.25 error={}".format(np_predictions_error[2]))
plt.scatter(np_feature, np_feature_predictions_parameters[:,3], label="mu=6 sigma=1.25 error={}".format(np_predictions_error[3]))
plt.legend(loc="upper right", bbox_to_anchor=(1.9, 1))
plt.xlabel("Feature", fontsize=16)
plt.ylabel("Response", fontsize=16)
```

```python
### Experiment 2: Multiple Linear Regression Model
#### Part I â€“ Given Data
```

```python
# Ingest advertising.csv and split the dataset by variables (column)
# The sales column is our dependent variable
np_offset = np.loadtxt(open('advertising.csv', "rb"), skiprows=1, delimiter=",", usecols=range(1,2))
np_tv = np.loadtxt(open('advertising.csv', "rb"), skiprows=1, delimiter=",", usecols=range(2,3))
np_radio = np.loadtxt(open('advertising.csv', "rb"), skiprows=1, delimiter=",", usecols=range(3,4))
np_newspaper = np.loadtxt(open('advertising.csv', "rb"), skiprows=1, delimiter=",", usecols=range(4,5))
np_sales = np.loadtxt(open('advertising.csv', "rb"), skiprows=1, delimiter=",", usecols=range(5,6))

# Generate 4 plots
fig, axs = plt.subplots(2, 2, dpi=80, figsize=(10,8))
axs[0,0].set_title("Offset vs Sales")
axs[0,0].scatter(np_offset, np_sales)
axs[0,0].set(xlabel='Offset', ylabel='Sales')

axs[0,1].set_title("TV vs Sales")
axs[0,1].scatter(np_tv, np_sales)
axs[0,1].set(xlabel='TV', ylabel='Sales')

axs[1,0].set_title("Radio vs Sales")
axs[1,0].scatter(np_radio, np_sales)
axs[1,0].set(xlabel='Radio', ylabel='Sales')

axs[1,1].set_title("Newspaper vs Sales")
axs[1,1].scatter(np_newspaper, np_sales)
axs[1,1].set(xlabel='Newspaper', ylabel='Sales')
fig.tight_layout() 
```

#### Part II â€“ Model Familiarity

Multiple Linear Regression  
$ y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \beta_3 x_{i3} + \beta_4 x_{i4} $

$ y_i $ = Sales  
$ \beta_0 $ = Constant  
$ x_1 $ = Offset   
$ \beta_1 $ = Offset coefficient  
$ x_2 $ = TV   
$ \beta_2 $ = TV coefficient  
$ x_3 $ = Radio   
$ \beta_3 $ = Radio coefficient  
$ x_4 $ = Newspaper   
$ \beta_4 $ = Newspaper coefficient  


#### Part III â€“ Model Implementation

```python
# Function that calculates the multiple linear regression
def multiple_linear_regression(np_b, np_x):
    sales = np_b[0] + np.multiply(np_b[1],np_x[1]) + np.multiply(np_b[2],np_x[2]) + np.multiply(np_b[3],np_x[3]) + np.multiply(np_b[4],np_x[4])
    return sales
```

#### Part IV â€“ Model output and Visualization + Part V â€“ Error between Model and Data:

```python
# Calculate predictions
np_b_1 = np.array([1.6,0,0.08,0.5,0.001])
np_b_2 = np.array([1.6,0,0.04,0.25,0.001])
np_x = np.array([np_offset, np_offset, np_tv, np_radio, np_newspaper])
np_sales_predictions_1 = multiple_linear_regression(np_b_1, np_x)
np_sales_predictions_2 = multiple_linear_regression(np_b_2, np_x)

# Calculate mean absolute error
np_predictions_error_1 = np.sum(abs(np_sales_predictions_1.T - np_sales))/np_sales.size
np_predictions_error_2 = np.sum(abs(np_sales_predictions_2.T - np_sales))/np_sales.size

print("Type 1 Coefficients: ")
print("\tConstant =", np_b_1[0])
print("\tOffset Coefficients =", np_b_1[1])
print("\tTV Coefficients =", np_b_1[2])
print("\tRadio Coefficients =", np_b_1[3])
print("\tNewspaper Coefficients =", np_b_1[4])
print("\tMean Absolute Error =", np_predictions_error_1)

print("\nType 2 Coefficients: ")
print("\tConstant =", np_b_2[0])
print("\tOffset Coefficients =", np_b_2[1])
print("\tTV Coefficients =", np_b_2[2])
print("\tRadio Coefficients =", np_b_2[3])
print("\tNewspaper Coefficients =", np_b_2[4])
print("\tMean Absolute Error =", np_predictions_error_2)


# Generate plots for predictions
fig_1, axs_1 = plt.subplots(2, 2, dpi=80, figsize=(10,8))
fig_1.suptitle("Prediction with Type 1 Coefficients")

axs_1[0,0].set_title("Offset vs Sales")
axs_1[0,0].scatter(np_offset, np_sales, label="Dataset")
axs_1[0,0].scatter(np_offset, np_sales_predictions_1, label="Prediction 1", alpha = 0.4)
axs_1[0,0].scatter(np_offset, np_sales_predictions_2, label="Prediction 2", alpha = 0.4)
axs_1[0,0].set(xlabel='Offset', ylabel='Sales')
axs_1[0,0].legend(loc="upper right")

axs_1[0,1].set_title("TV vs Sales")
axs_1[0,1].scatter(np_tv, np_sales, label="Dataset")
axs_1[0,1].scatter(np_tv, np_sales_predictions_1, label="Prediction 1", alpha = 0.4)
axs_1[0,1].scatter(np_tv, np_sales_predictions_2, label="Prediction 2", alpha = 0.4)
axs_1[0,1].set(xlabel='TV', ylabel='Sales')
axs_1[0,1].legend(loc="upper right")

axs_1[1,0].set_title("Radio vs Sales")
axs_1[1,0].scatter(np_radio, np_sales, label="Dataset")
axs_1[1,0].scatter(np_radio, np_sales_predictions_1, label="Prediction 1", alpha = 0.4)
axs_1[1,0].scatter(np_radio, np_sales_predictions_2, label="Prediction 2", alpha = 0.4)
axs_1[1,0].set(xlabel='Radio', ylabel='Sales')
axs_1[1,0].legend(loc="upper right")

axs_1[1,1].set_title("Newspaper vs Sales")
axs_1[1,1].scatter(np_newspaper, np_sales, label="Dataset")
axs_1[1,1].scatter(np_newspaper, np_sales_predictions_1, label="Prediction 1", alpha = 0.4)
axs_1[1,1].scatter(np_newspaper, np_sales_predictions_2, label="Prediction 2", alpha = 0.4)
axs_1[1,1].set(xlabel='Newspaper', ylabel='Sales')
axs_1[1,1].legend(loc="upper right")

fig_1.tight_layout() 
```
